```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_descriptor  = pd.read_csv('descriptor_data.csv')
df_descriptor = df_descriptor.drop(['Unnamed: 0', 'index', 'SPS.1'], axis=1)
df_descriptor
```

```python
target_values = pd.read_excel('input-Tg-DSC.xlsx')
target_values = target_values['Tg(℃)']
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

std = MinMaxScaler()
df_descriptor_s = std.fit_transform(df_descriptor)
```

```python
from sklearn.feature_selection import VarianceThreshold
import pandas as pd

# 假设X为特征矩阵，columns为特征列标签
sel = VarianceThreshold(threshold=0.01)
X_sel = sel.fit_transform(df_descriptor_s)
feature_index = sel.get_support()

# 获取选择的特征列标签
selected_columns = df_descriptor.columns[feature_index]

# 创建包含选择特征的DataFrame
X_sel_df = pd.DataFrame(X_sel, columns=selected_columns)

print('number of features after remove redundant ones:', sum(feature_index))
```

```python
y = target_values
```

# 特征重要性

```python
# 导入必要的库
from catboost import CatBoostRegressor
from sklearn.feature_selection import SelectFromModel
import pandas as pd


# 初始化CatBoost回归模型
catb = CatBoostRegressor()
# iterations=1000, learning_rate=0.1, depth=6, loss_function='RMSE'
# 拟合模型
catb.fit(X_sel_df, y, verbose=False)

# 使用特征重要性进行特征选择
feature_importances = catb.get_feature_importance()
threshold = 1  # 设置特征重要性阈值
selected_features = X_sel_df.columns[feature_importances > threshold]

# 仅保留重要特征
X_selected = X_sel_df[selected_features]
X_selected
```

```python
# 将归一化数据还原为原数据
X_selected.columns
```

```python
X_selected = df_descriptor[['qed', 'MaxPartialCharge', 'MaxAbsPartialCharge', 'BCUT2D_MWLOW',
       'BCUT2D_CHGHI', 'BCUT2D_MRHI', 'BertzCT', 'PEOE_VSA6', 'PEOE_VSA8',
       'PEOE_VSA9', 'SMR_VSA6', 'SMR_VSA7', 'SlogP_VSA1', 'SlogP_VSA11',
       'SlogP_VSA4', 'SlogP_VSA8', 'VSA_EState2', 'VSA_EState8',
       'FractionCSP3', 'NumAromaticRings', 'NumHAcceptors',
       'NumRotatableBonds', 'RingCount', 'fr_aniline', 'fr_bicyclic',
       'fr_ether']]
X_selected
```

```python
# 将dataframe存为csv文件
X_selected.to_csv('Feature_importance_catb_26.csv', index=False)
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2,random_state=1)

# 初始化CatBoost回归模型
catb = CatBoostRegressor()

# 拟合模型
catb.fit(X_train, y_train, eval_set=(X_test, y_test), verbose=False)
```

```python
# 进行预测
y_pred_train_catb = catb.predict(X_train)
y_pred_test_catb = catb.predict(X_test)

plt.figure(figsize=(4,4))
plt.plot([0, 500], [0, 500], 'k--', label = 'y = x')
plt.scatter(y_train, y_pred_train_catb, alpha=0.9, label='Training set', color='cornflowerblue', marker='o', s=18)  # Set color to cornflowerblue
plt.scatter(y_test, y_pred_test_catb, alpha=0.8, label='Test set', color='tomato', marker='^', s=18)  # Set color to tomato
plt.xlim(0, 500)
plt.ylim(0, 500)
plt.title('CatBoost')
plt.legend()
plt.xlabel('Experimental(℃)')
plt.ylabel('Predicted(℃)')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score , mean_absolute_error, make_scorer

print('training R = %.3f' % np.sqrt(r2_score(y_train, y_pred_train_catb)))
print('training R2 = %.3f' % r2_score(y_train, y_pred_train_catb))
print('training RMSE = %.3f' % np.sqrt(mean_squared_error(y_train, y_pred_train_catb)))

print('test R = %.3f' % np.sqrt(r2_score(y_test, y_pred_test_catb)))
print('test R2 = %.3f' % r2_score(y_test, y_pred_test_catb))
print('test RMSE = %.3f' % np.sqrt(mean_squared_error(y_test, y_pred_test_catb)))
print('test MAE = %.3f' %mean_absolute_error(y_test, y_pred_test_catb))
```

# 误差分布

```python
plt.figure(figsize=(5, 4))
plt.hist(best_catb.predict(X_train)-y_train, color='cornflowerblue',
         bins=np.arange(-100, 100, 2), density=True, label= 'Training set' )
plt.hist(best_catb.predict(X_test)-y_test, color='tomato',
         bins=np.arange(-100, 100, 2), density=True, alpha=0.6, label= 'Test set' )
plt.legend()
plt.savefig('CATB-误差分布-1.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
# 计算绝对误差
train_errors = np.abs(best_catb.predict(X_train) - y_train)
test_errors = np.abs(best_catb.predict(X_test) - y_test)

# 定义误差范围
范围1 = (0, 20)
范围2 = (20, 40)
范围3 = (40, 60)
范围4 = (60, 80)
范围5 = (80, 100)

# 统计落入每个范围内的误差数量
train_range1_count = np.sum((train_errors >= 范围1[0]) & (train_errors < 范围1[1]))
train_range2_count = np.sum((train_errors >= 范围2[0]) & (train_errors < 范围2[1]))
train_range3_count = np.sum((train_errors >= 范围3[0]) & (train_errors < 范围3[1]))
train_range4_count = np.sum((train_errors >= 范围4[0]) & (train_errors < 范围4[1]))
train_range5_count = np.sum((train_errors >= 范围5[0]) & (train_errors < 范围5[1]))
test_range1_count = np.sum((test_errors >= 范围1[0]) & (test_errors < 范围1[1]))
test_range2_count = np.sum((test_errors >= 范围2[0]) & (test_errors < 范围2[1]))
test_range3_count = np.sum((test_errors >= 范围3[0]) & (test_errors < 范围3[1]))
test_range4_count = np.sum((test_errors >= 范围4[0]) & (test_errors < 范围4[1]))
test_range5_count = np.sum((test_errors >= 范围5[0]) & (test_errors < 范围5[1]))

# 创建柱状图
plt.figure(figsize=(4, 4))
plt.bar([0.75, 2.75, 4.75, 6.75, 8.75], [train_range1_count, train_range2_count, train_range3_count, train_range4_count, train_range5_count],
        width=0.5, color='cornflowerblue', edgecolor = 'black')
plt.bar([1.25, 3.25, 5.25, 7.25, 9.25], [test_range1_count, test_range2_count, test_range3_count, test_range4_count, test_range5_count], 
        width=0.5, color='tomato',edgecolor = 'black', alpha=0.7)
plt.xticks([1, 3, 5, 7, 9], ['0-20', '20-40', '40-60', '60-80', '80-100'])
# plt.ylabel('Number of PIs')
# plt.xlabel('Absolute error range')

# 去掉上和右边框
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.xticks(rotation=-45)
plt.yticks(rotation=-90)
plt.savefig('CATB-误差分布-2.png',dpi=300, bbox_inches='tight')
plt.show()
```

# William plot

```python
import numpy as np
import matplotlib.pyplot as plt
from catboost import CatBoostRegressor

# 假设你已经有一个构建好的CatBoost预测模型

# 假设 best_catb 是你的CatBoost模型
# 假设 X_train 是你的训练特征数据
# 假设 y_train 是你的训练目标变量
# 假设 X_test 是你的测试特征数据

# 获取模型的预测值
y_pred_train_catb = best_catb.predict(X_train)
y_pred_test_catb = best_catb.predict(X_test)

# 计算训练数据的残差
residuals_train = y_train - y_pred_train_catb

# 计算训练数据的标准化残差
std_residuals_train = residuals_train / np.std(residuals_train)

# 绘制预测值和残差的散点图
plt.figure(figsize=(8, 6))
plt.scatter(y_pred_train_catb, residuals_train, color='blue')
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.grid(True)
plt.show()

# 绘制标准化残差的分布图
plt.figure(figsize=(8, 6))
plt.hist(std_residuals_train, bins=20, color='blue', edgecolor='black')
plt.xlabel('Standardized Residuals')
plt.ylabel('Frequency')
plt.title('Standardized Residuals Distribution')
plt.grid(True)
plt.show()
```
